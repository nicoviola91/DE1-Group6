## create pandas DataFrame

````python
# Create the pandas DataFrame with column name is provided explicitly

columns=['Song_title','Artist','Year','Hotness_of_song','Key','Tempo']
main_directory = 'MillionSongSubset'

coldict = {i:[] for i in columns}

for root, dirs, files in os.walk(main_directory):
    for filename in files:
        f = h5py.File(os.path.join(root, filename), 'r')
        songs_musicbrainz = f['musicbrainz']['songs']
        songs_metadata = f['metadata']['songs']
        songs_analysis = f['analysis']['songs']
        
        if songs_musicbrainz['year'][0]:
            coldict['Year'].append(songs_musicbrainz['year'][0])

            coldict['Song_title'].append(songs_metadata['title'][0].decode('UTF-8'))
            coldict['Artist'].append(songs_metadata['artist_name'][0].decode('UTF-8'))
            coldict['Hotness_of_song'].append(songs_metadata['song_hotttnesss'][0])

            coldict['Tempo'].append(songs_analysis['tempo'][0])
            coldict['Key'].append(songs_analysis['key'][0])
        

df = pd.DataFrame(coldict)        
df = df.convert_dtypes()
df['Hotness_of_song'] = df['Hotness_of_song'].fillna(-0.1)
````

## Plot a bit

````python
df.plot(x ='Year', y='Hotness_of_song', kind='scatter')
````

````python
df['Year'].hist(bins=30, edgecolor='black', range=[1927, 2010], color='purple')
plt.title('Number of songs over the years')
plt.xlabel('Year')
plt.ylabel('Nr of songs')
````

````python
def mean(a):
    b = list(a)
    if len(b):
        return sum(b)/len(b)
    return np.nan
````

````python
num_keys = 4 # out of 12

fig, ax = plt.subplots(1,1)

years_range = range(1927, 2010)

keys = list(set(df['Key']))[:num_keys]

for index, key_i in enumerate(keys):
    res_mean = []
    for anno in years_range:
        res_mean.append(mean(df.query(f'Year=={anno} and Key=={key_i}')['Hotness_of_song']))
    mask = np.isfinite(res_mean)
    ax.plot(np.array(years_range)[mask], np.array(res_mean)[mask], label=key_i)

ax.set(xlabel='Year', ylabel='Hotness', xlim=[years_range[0],years_range[-1]])
plt.legend()
    
````

````python
num_keys = 4 # out of 12

fig, ax = plt.subplots(1,1)

years_range = range(1927, 2010)

keys = list(set(df['Key']))[:num_keys]
q_keys = []
for index, key_i in enumerate(keys):
    q_keys.append(df.query(f'Key=={key_i}')['Year'])
ax.hist(q_keys, bins=30, range=[years_range[0],years_range[-1]], label=keys[:num_keys])

ax.set(xlabel='Year', ylabel='Frequency', xlim=[years_range[0],years_range[-1]])
plt.legend()
    
````

## Spark

````python
from operator import add
from pyspark.sql import SparkSession

def sort_1(e):
    return e[0]

def sort_2(e):
    return e[1]

spark = SparkSession.builder.appName(
  "pandas to spark").getOrCreate()
````

````python
sdf = spark.createDataFrame(df)
````

````python
sdf.show(10)
````

````python
year_freq = sorted(sdf.select('Year').rdd\
                .map(lambda x: (x, 1))\
                .reduceByKey(add).collect(), key=sort_1, reverse=True)
years = [i[0][0] for i in year_freq]
years_freq_ =[i[1] for i in year_freq]
````

````python
plt.plot(years, years_freq_,'o', color='purple')
plt.title('Number of songs over the years')
plt.xlabel('Year')
plt.ylabel('Nr of songs')
````

````python
year_freq = sorted(sdf.where('Key==0').select('Year').rdd\
                .map(lambda x: (x, 1))\
                .reduceByKey(add).collect(), key=sort_1, reverse=True)
years = [i[0][0] for i in year_freq]
years_freq_ =[i[1] for i in year_freq]
plt.plot(years, years_freq_,'o', color='purple')
plt.title('Number of songs over the years with key: 0')
plt.xlabel('Year')
plt.ylabel('Nr of songs')
````
